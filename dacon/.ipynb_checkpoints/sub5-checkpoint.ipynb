```python
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras.utils import to_categorical
```

```python
X_train_full = (train[[str(i) for i in range(784)]]/255.).values.reshape(-1,28,28,1)
y_train_full = to_categorical(train['digit'].values)
```

```python
def adjust(value, threshold):
    value = value -threshold
    value = np.array([0 if i < 0 else i for i in value])
    return value.reshape(28,28)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(
    rotation_range = 36,
    zoom_range = 0.1,
    width_shift_range = 0.1,
    height_shift_range = 0.1,
    validation_split=0.10
)
```

```python
batch_size = 8
```

```python
train_generator = datagen.flow(
    X_train_full, y_train_full,
    batch_size=batch_size,
    subset='training'
)
validation_generator = datagen.flow(
    X_train_full, y_train_full,
    batch_size=batch_size,
    subset='validation'
)
```

```python
input_ = layers.Input(shape=(28,28,1))
x = layers.Conv2D(32, 3, strides=2, padding="same")(input_)
x = layers.BatchNormalization()(x)
x = layers.LeakyReLU(alpha=0.2)(x)

x = layers.Conv2D(64, 3, padding="same")(x)
x = layers.BatchNormalization()(x)
x = layers.LeakyReLU(alpha=0.2)(x)
    
previous_block_activation = x
    
for size in [128, 256]:
    x = layers.LeakyReLU(alpha=0.2)(x)
    x = layers.SeparableConv2D(size, 3, padding="same")(x)
    x = layers.BatchNormalization()(x)
        
    x = layers.LeakyReLU(alpha=0.2)(x)
    x = layers.SeparableConv2D(size, 3, padding="same")(x)
    x = layers.BatchNormalization()(x)
        
    x = layers.MaxPooling2D(3, strides=2, padding="same")(x)
        
    residual = layers.Conv2D(size, 1, strides=2, padding="same")(
        previous_block_activation        
    )
    x = layers.add([x, residual])
    previous_block_activation = x
        
x = layers.SeparableConv2D(128, 3, padding="same")(x)
x = layers.BatchNormalization()(x)
x = layers.LeakyReLU(alpha=0.2)(x)
    
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.5)(x)
output = layers.Dense(10, activation="softmax")(x)

model = models.Model(inputs=[input_], outputs=[output])
```

```python
model.summary()
```

```python
model.compile(optimizer="rmsprop", loss="categorical_crossentropy", 
             metrics=["accuracy"])
```

```python
epochs = 100
history = model.fit_generator(
    train_generator, 
    epochs = epochs,
    validation_data = validation_generator)
```

```python
print(
    f"Train accuracy={max(history.history['accuracy']):.5f}, " +
    f"Validation accuracy={max(history.history['val_accuracy']):.5f}"
)
```

```python
plt.figure()
plt.plot(history.history['accuracy'],label='accuracy')
plt.plot(history.history['val_accuracy'],label='validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
