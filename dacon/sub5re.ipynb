```python
import pandas as pd
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
submission = pd.read_csv('./submission.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
```

```python
np.random.seed(42)
tf.random.set_seed(42)
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras.utils import to_categorical
```

```python
threshold = 70
X = (train[[str(i) for i in range(784)]]).values
X = X- threshold
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        if X[i][j] < 0:
            X[i][j] = 0
```

```python
X_train_full = (X/255.).reshape(-1,28,28,1)
y_train_full = to_categorical(train['digit'].values)
```

```python
X_train_full = (train[[str(i) for i in range(784)]]/255.).values.reshape(-1,28,28,1)
y_train_full = to_categorical(train['digit'].values)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
    rotation_range = 36,
    zoom_range = 0.10,
    width_shift_range = 0.1,
    height_shift_range = 0.1,
    validation_split=0.10
)
```

```python
batch_size = 32
train_generator = datagen.flow(
    X_train_full, y_train_full,
    batch_size=batch_size,
    subset='training'
)
validation_generator = datagen.flow(
    X_train_full, y_train_full,
    batch_size=batch_size,
    subset='validation'
)
```

```python
input_2 = layers.Input(shape=(28,28,1))
x = layers.Conv2D(32, kernel_size=3, padding="same", activation="relu")(input_2)
x = layers.Conv2D(32, kernel_size=3, padding="same", activation="relu")(x)
x = layers.Conv2D(32, kernel_size=3, padding="valid",strides=2, activation="relu")(x)
#x = layers.MaxPooling2D(pool_size=(2,2))(x)
x = layers.Dropout(0.5)(x)

x = layers.Conv2D(64, kernel_size=3, padding="same", activation="relu")(x)
x = layers.Conv2D(64, kernel_size=3, padding="same", activation="relu")(x)
x = layers.Conv2D(64, kernel_size=3, padding="valid",strides=2, activation="relu")(x)
#x = layers.MaxPooling2D(pool_size=(2,2))(x)
x = layers.Dropout(0.5)(x)

x = layers.Conv2D(128, kernel_size=3, padding="same", activation="relu")(x)
x = layers.Conv2D(128, kernel_size=3, padding="same", activation="relu")(x)
x = layers.Conv2D(128, kernel_size=3, padding="valid",strides=2, activation="relu")(x)
#x = layers.MaxPooling2D(pool_size=(2,2))(x)
x = layers.Dropout(0.5)(x)

x = layers.Conv2D(128, kernel_size=3, padding="same", activation="relu")(x)
x = layers.Flatten()(x)
x = layers.Dropout(0.5)(x)
output_2 = layers.Dense(10, activation = "softmax")(x)

base_model2 = models.Model(inputs=[input_2], outputs=[output_2])
```

```python
activation=layers.LeakyReLU(alpha=0.2)
input_ = layers.Input(shape=(28,28,1))
x = layers.Conv2D(64, kernel_size=3, padding="same", activation=activation)(input_)
x = layers.Conv2D(64, kernel_size=3, padding="same", activation=activation)(x)
x = layers.Conv2D(64, kernel_size=3, padding="same", activation=activation)(x)
x = layers.Conv2D(64, kernel_size=3, padding="valid",strides=2, 
                  activation=activation)(x)
x = layers.Dropout(0.5)(x)

x = layers.Conv2D(128, kernel_size=3, padding="same", activation=activation)(x)
x = layers.Conv2D(128, kernel_size=3, padding="same", activation=activation)(x)
x = layers.Conv2D(128, kernel_size=3, padding="same", activation=activation)(x)
x = layers.Conv2D(128, kernel_size=3, padding="valid",strides=2, 
                  activation=activation)(x)
x = layers.Dropout(0.5)(x)

x = layers.Conv2D(256, kernel_size=3, padding="same", activation=activation)(x)
x = layers.Conv2D(256, kernel_size=3, padding="same", activation=activation)(x)
x = layers.Conv2D(256, kernel_size=3, padding="same", activation=activation)(x)
x = layers.Conv2D(256, kernel_size=3, padding="valid",strides=2, 
                  activation=activation)(x)
x = layers.Dropout(0.5)(x)

#x = layers.Conv2D(256, kernel_size=3, padding="same", activation=activation)(x)
x = layers.Flatten()(x)
x = layers.Dropout(0.5)(x)
output_ = layers.Dense(10, activation = "softmax")(x)

base_model = models.Model(inputs=[input_], outputs=[output_])
```

```python
base_model.summary()
```

```python
model = base_model
model.compile(optimizer="rmsprop", loss="categorical_crossentropy", 
             metrics=["accuracy"])
```

```python
epochs = 500
history = model.fit_generator(
    train_generator, 
    epochs = epochs,
    validation_data = validation_generator)
```

```python
print(
    f"Train accuracy={max(history.history['accuracy']):.5f}, " +
    f"Validation accuracy={max(history.history['val_accuracy']):.5f}"
)
```

```python
plt.figure()
plt.plot(history.history['accuracy'],label='accuracy')
plt.plot(history.history['val_accuracy'],label='validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
